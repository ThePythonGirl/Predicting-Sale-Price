## Before start modellig we need to do the following:
1. Convert string to numbers
2. Fill missing values

we can turn all of our data into numbers is by converting them into pandas catgories.

**convert string into categories using pandas api**

```python
# we can check for string type using pandas api
pd.api.types.is_string_dtype(df["UsageBand"])
```

```python
# can use the above api to find the columns which contains strings
for label, content in df.items():
    if pd.api.types.is_string_dtype(content):
        print(label)
```

```python
# can use the above and turn them into categories
for label, content in df.items():
    if pd.api.types.is_string_dtype(content):
        df[label] = content.astype("category").cat.as_ordered()
```

```python
df.info()
```

```python
df.state.cat.codes
```

**Filling missing values in the numeric columns only**

```python
# check for columns having numeric values
for label, content in df.items():
    if pd.api.types.is_numeric_dtype(content):
        print(label)
```

```python
# check for columns having null values
for label, content in df.items():
    if pd.api.types.is_numeric_dtype(content):
        if pd.isnull(content).sum():
            print(label)
```

```python
# using the above and then fill numeric rows with the median 
for label, content in df.items():
    if pd.api.types.is_numeric_dtype(content):
        if pd.isnull(content).sum():
            df[label] = content.fillna(content.median())
```

```python
df.isna().sum()
```

**Filling and turning categorical variables into numbers**

```python
# Check for columns which are not numeric
for label, content in df.items():
    if not pd.api.types.is_numeric_dtype(content):
        print(label)
```

```python
# using above and turning categorical variables into numbers and filling them
for label, content in df.items():
    if not pd.api.types.is_numeric_dtype(content):
        df[label] = pd.Categorical(content).codes+1
```

**RandomForestRegressor Model**

```python
%%time
from sklearn.ensemble import RandomForestRegressor

np.random.seed(42)

model = RandomForestRegressor()

model.fit(df.drop("SalePrice", axis=1), df["SalePrice"])
```

```python
# score the model
model.score(df.drop("SalePrice", axis=1), df["SalePrice"])
```
